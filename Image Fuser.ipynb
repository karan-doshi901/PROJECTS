```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as p

a=input("enter First Image:")
b=input("enter second Image:")

# Open the two images to be fused
img1 = Image.open(a)
img2 = Image.open(b)

c=int(input("enter size:"))
img1=img1.resize((c,c))
img2=img2.resize((c,c))

# Convert the images to numpy arrays
img1_np = np.array(img1)
img2_np = np.array(img2)

mini=np.zeros((c,c,3))
max1=np.zeros((c,c,3))
avg1=np.zeros((c,c,3))

for i in range(0,c):
  for j in range(0,c):
    for k in range(3):
      a=img1_np[i,j,k]
      b=img2_np[i,j,k]
      mini[i,j,k]=min(a,b)
      max1[i,j,k]=max(a,b)
      avg1[i,j,k]=(a+b)//2


print("SELECT 1 FOR MINIMUM, SELECT 2 FOR MAXIMUM,SELECT 3 FOR AVERAGE, SELECT 4 TO SEE ALL , SELECT 5 TO QUIT")
x=int(input("Enter: "))
while x!=5:
  if x==1:
    p.subplot(1,3,1)
    p.imshow(img1)

    p.subplot(1,3,2)
    p.imshow(img2)

    p.subplot(1,3,3)
    fused_img = Image.fromarray(np.uint8(mini))
    photo=p.imshow(fused_img)
    print(photo)



  elif x==2:
    p.subplot(1,3,1)
    p.imshow(img1)

    p.subplot(1,3,2)
    p.imshow(img2)

    p.subplot(1,3,3)

    fused_img = Image.fromarray(np.uint8(max1))
    photo=p.imshow(fused_img)
    print(photo)


  elif x==3:
    p.subplot(1,3,1)
    p.imshow(img1)

    p.subplot(1,3,2)
    p.imshow(img2)

    p.subplot(1,3,3)

    fused_img = Image.fromarray(np.uint8(avg1))
    photo=p.imshow(fused_img)
    print(photo)

#------------------
  elif x==4:
    p.subplots_adjust(wspace=1,hspace=1)

    p.subplot(2,2,1)

    fused_img = Image.fromarray(np.uint8(mini))
    p.imshow(fused_img)
    p.xlabel("MINIMUM")

    p.subplot(2,2,2)

    fused_img = Image.fromarray(np.uint8(max1))
    p.imshow(fused_img)
    p.xlabel("MAXIMUM")


    p.subplot(2,2,3)

    fused_img = Image.fromarray(np.uint8(avg1))
    p.imshow(fused_img)
    p.xlabel("AVERAGE")
  
  x=int(input("to stop:"))






















```
